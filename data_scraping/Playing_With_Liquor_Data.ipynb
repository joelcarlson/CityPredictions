## Convert date columns into reasonable formats

```python
check_orig = nyc['License Original Issue Date'].copy()
check_eff = nyc['License Effective Date'].copy()
check_exp = nyc['License Expiration Date'].copy()
print len(check_orig[pd.isnull(check_orig)]), len(check_eff[pd.isnull(check_eff)]), len(check_exp[pd.isnull(check_exp)])
```

Based on the above, we will convert the NULL values in the original issue date to their effective date

```python
nyc['License Original Issue Date'] = np.where(pd.isnull(nyc['License Original Issue Date']), nyc['License Effective Date'],
                                              nyc['License Original Issue Date'])
```

There is one comically broken value in license effective

```python
nyc.loc[nyc['License Effective Date'] == "10/11/0130", 'License Effective Date'] = "10/11/2007"
```

```python
nyc.loc[nyc['License Expiration Date'] == "28/02/3017", 'License Expiration Date'] = "02/28/2017"
```

```python
import time
start_time = time.time()
nyc['License_Original_Issue_Date'] = pd.to_datetime(nyc['License Original Issue Date'])
print "Originals converted", time.time() - start_time
nyc['License_Effective_Date'] = pd.to_datetime(nyc['License Effective Date'])
print "Effectives converted", time.time() - start_time
nyc['License_Expiration_Date'] = pd.to_datetime(nyc['License Expiration Date'])
print "Expirations converted", time.time() - start_time
```
